```python
import gzip
import json
import pandas as pd

META_PATH = "/content/drive/MyDrive/data/meta_Electronics.jsonl.gz"
OUTPUT_PATH = "/content/drive/MyDrive/data/asin_image_map.csv"

rows = []

with gzip.open(META_PATH, "rt", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)

        asin = obj.get("parent_asin")
        images = obj.get("images", [])

        if asin and images:
            img = images[0].get("hi_res") or images[0].get("large")
            if img:
                rows.append({
                    "asin": asin,
                    "image": img
                })

image_df = pd.DataFrame(rows)
image_df.to_csv(OUTPUT_PATH, index=False)

print("‚úÖ ASIN ‚Üí Image mapping saved to:", OUTPUT_PATH)
print("üñºÔ∏è Total products with images:", len(image_df))
image_df.head()
```
