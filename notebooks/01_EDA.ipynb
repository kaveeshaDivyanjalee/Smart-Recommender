```python
import pandas as pd

df = pd.read_csv(
    '../data/Electronics.csv.gz',
    compression='gzip'
)

df.head()
```

```python
#EDA-Exploratory Data Analysis
#1-Quick sanity check
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#2:Rename columns
df.columns
```

```python
df = df.rename(columns={
    'user_id': 'user',
    'parent_asin': 'item',
    'rating': 'rating',
    'timestamp': 'timestamp'
})
```

```python
df.head()
```

```python
#3: Basic statistics
n_users = df['user'].nunique()
n_items = df['item'].nunique()
n_interactions = len(df)

print(f"Users: {n_users}")
print(f"Items: {n_items}")
print(f"Interactions: {n_interactions}")
print(f"Sparsity: {1 - n_interactions/(n_users*n_items):.5f}")
```

```python
#4: Rating distribution
import matplotlib.pyplot as plt

df['rating'].value_counts().sort_index().plot(kind='bar')
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

```

```python
#5: User activity distribution
user_counts = df.groupby('user').size()

user_counts.describe()
```

```python
user_counts.hist(bins=50)
plt.title("Reviews per User")
plt.show()
```

```python
#6: Item popularity distribution
item_counts = df.groupby('item').size()

item_counts.describe()
```

```python
item_counts.hist(bins=50)
plt.title("Reviews per Item")
plt.show()
```

```python
#7: Time analysis (if timestamp exists)
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')

df['year'] = df['timestamp'].dt.year
df['year'].value_counts().sort_index().plot()
plt.title("Interactions over time")
plt.show()
```
